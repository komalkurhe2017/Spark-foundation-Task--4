# d. Data Visualisation

```python
# Let's see in which year, attacks happened most number of times

plt.rcParams['figure.figsize']=15,11
sns.countplot(y=df['iyear'], palette='inferno')

plt.xlabel('Number of attacks', fontsize=20, fontweight='bold')
plt.ylabel('Attack Year', fontsize=20, fontweight='bold')

plt.title('Trend of increase in Terror from 1970 to 2017', fontweight="bold", size=25)

plt.xticks(fontweight="bold", size=15)
plt.yticks(fontweight="bold", size=15)

plt.show() 
```

```python
plt.figure(figsize=(15,10))
sns.countplot(x='iyear',hue='success',data=df)

plt.xlabel('Attack Year', fontsize=20, fontweight='bold')
plt.ylabel('Number of attacks', fontsize=20, fontweight='bold')

plt.title('Trend of increase in Terror from 1970 to 2017', fontweight="bold", size=25)

plt.xticks(rotation=60, fontweight="bold", size=15)
plt.yticks(fontweight="bold", size=15)

plt.show()
```

```python
df['region_txt'].value_counts()
```

```python
#let's check which region is affected by terrorism in each year
@interact

def attack_year(year=list(df['iyear'].unique())):
    a = df[df['iyear']==year]
    return sns.countplot(y=a['region_txt'], palette='viridis')
```

```python
#let's check which countries are affected more from each region
@interact
def country_name(region=list(df['region'].unique())):
    a=df[(df['region']==region)]
    a=pd.DataFrame(a['country'].value_counts())
    return a.style.background_gradient(cmap='plasma')
```

```python
#let's check which cities attacked most of the time from each country
@interact
def city_name(country=list(df['country'].unique())):
    a=df[df['country']==country]
    a=pd.DataFrame(a['city'].value_counts().head(10))
    return a.style.background_gradient(cmap='cool')
```

```python
plt.figure(figsize=(15,6))

sns.countplot(x='region_txt',data=df)

plt.xlabel('Region-wise', fontsize=20, fontweight='bold')
plt.ylabel('Number of attacks', fontsize=20, fontweight='bold')

plt.title('Total number of attacks Region-wise', fontweight="bold", size=25)

plt.xticks(rotation=90, fontweight="bold", size=15)
plt.yticks(fontweight="bold", size=15)

plt.show()
```
